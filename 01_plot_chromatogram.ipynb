```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline

#This is a code to generate chromatograms for the predicted retention time
def plot_chromatograms(array_rt=[3,5,6], name=["a","b","c"], end_time_of_lc_run=10, title='Reverse Phase Chromatography'):

    baseline_time = [round(i,2) for i in np.arange(0, end_time_of_lc_run, 0.01).tolist()]
    
    all_the_peaks = []
    for rt in array_rt:
        baseline_time += [np.round(i,2) for i in np.random.normal(rt, 0.05, 2000)]#5000
    
    num_most_common = Counter(baseline_time).most_common(1)[0][1] 
    
    fig, ax = plt.subplots(figsize=(20,3))
    #sns.set(color_codes=True)
    sns.set_style("whitegrid")
    ax.set_xlim(0,end_time_of_lc_run)
    ax.grid(False)
    for i, n in zip (array_rt, name):
        ax.text(i, num_most_common*1.2, n, fontsize=20, rotation=450)

    chrom = sns.distplot(baseline_time,hist=True,kde=False,bins=500, color="green",kde_kws={"bw":0.03} )
    chrom.set_ylabel("Intensity",fontsize=20)
    chrom.set_xlabel("min",fontsize=20)
    chrom.axes.set_title(title,fontsize=20)
    chrom.tick_params(labelsize=20)
    plt.show()

target_name = ["pep_A", "pep_B", "pep_C", "pep_D", "pep_E", "pep_F", "pep_G"]
end_time = 10

plot_chromatograms(array_rt=[3,5,6,5,3.4,2.4,8.9], name = target_name, end_time_of_lc_run=end_time, title='Reverse Phase Chromatography')
plot_chromatograms(array_rt=[1,3,4,8,9.4,8.4,4.9], name = target_name, end_time_of_lc_run=end_time, title='Strong Cation Chromatography')
plot_chromatograms(array_rt=[2.1,3,6,4,8.4,6.1,3.2], name = target_name, end_time_of_lc_run=end_time, title='HILIC')
```
